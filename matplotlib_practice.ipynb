```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()
fig = plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label='Boston')
#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45.
plt.ylim(0,45)
#Create a title.
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, label='Boston',color="green",linewidth=2,marker="*")
#Create a Title
ax.set_title("PyBer Fare by Month")
#Set labels for axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#Add a legend
ax.legend()
#Add gridlines 
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,"o")
```

```python
plt.scatter(y_axis,x_axis,color='red',label='Chicago')
#Add a title
plt.title("PyBer Fare by Month")
#Label the X Axis
plt.xlabel("Fare($)")
#Label the Y Axis
plt.ylabel("Date")
#Add a legend
plt.legend()
```

```python
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,color="skyblue",s=[i*5 for i in y_axis],alpha=.8,edgecolors="black",linewidth=2,label="Boston")
#Add a title
ax.set_title("PyBer Fare by Month")
#Add Axes Titles
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
#Add Legend
ax.legend()
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
#Assign 12 colors, one for each month.
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
explode_values=(0,0,0,0,0,0,.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values,colors=colors,labels=x_axis,autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
explode_values=(0,0,.2,0,0,0,.2,0,0,0,0,0)
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,explode=explode_values,colors=colors,autopct='%.1f%%',shadow=True,startangle=90,counterclock=False)
plt.show()
```
